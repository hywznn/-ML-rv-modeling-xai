## 범주형 변수 데이터 탐색

```python
from sklearn.preprocessing import LabelEncoder

# 범주형 피처 리스트
categorical_cols = [
    'VIP등급코드', '최상위카드등급코드',
    '거주시도명', 'Life_Stage', '카드론동의여부'
]

# 범주형 피처 정보 출력
for col in categorical_cols:
    print(f"\n📌 [ {col} ]")
    print(f"- 고유값 개수: {df_cleaned[col].nunique()}")
    print(f"- 고유값 목록 (앞 10개): {df_cleaned[col].unique()[:10]}")
    print(f"- 각 범주별 개수 (상위 10개):\n{df_cleaned[col].value_counts().head(10)}")

# 요약 통계 테이블 출력
print("\n📊 범주형 피처 전체 요약 통계:")
print(df_cleaned[categorical_cols].describe(include='all').T)
```

```python
# 최상위등급코드 결측치 많음( _ 가 많음) -> 제거
# 거주시도명 -> 원핫인코딩해서 자료 정리
# Life_Stage -> 원핫인코딩해서 자료 정리
# 카드론동의여부 -> 원핫인코딩해서 자료 정리 ( 0,1 임 )
# VIP등급코드 -> 결측치 많음 ( _ 가 많음) -> 제거
```

```python
# 1. '_' → 결측 처리 후 제거할 필요는 없지만, 바로 제거
df_cleaned.drop(columns=['VIP등급코드', '최상위카드등급코드'], inplace=True)

# 2. 카드론동의여부: 'Y' → 1, 'N' → 0
df_cleaned['카드론동의여부'] = df_cleaned['카드론동의여부'].map({'Y': 1, 'N': 0})
```

```python
# 1. 순서 정의
시도_순서 = [
    '서울', '경기', '인천',
    '부산', '대구', '광주', '대전', '울산', '세종',
    '강원', '충북', '충남', '전북', '전남', '경북', '경남', '제주'
]

life_stage_order = [
    '1.Single',
    '2.가족형성기',
    '3.자녀출산기',
    '4.자녀성장기(1)',
    '5.자녀성장기(2)',
    '6.자녀출가기',
    '7.노령'
]

# 2. 매핑 딕셔너리 생성
city_map = {name: idx for idx, name in enumerate(시도_순서)}
life_map = {name: idx for idx, name in enumerate(life_stage_order)}

# 3. 실제 라벨 인코딩 적용
df_cleaned['거주시도명'] = df_cleaned['거주시도명'].map(city_map)
df_cleaned['Life_Stage'] = df_cleaned['Life_Stage'].map(life_map)

# 결과 확인
print("✅ 라벨 인코딩 완료:")
print(df_cleaned[['거주시도명', 'Life_Stage']].head())
```

```python
# 시도별 순서 정의 및 라벨 부여
# 라벨 번호 기준:
# 0: 서울, 1: 경기, 2: 인천, 3: 부산, 4: 대구, 5: 광주, 6: 대전,
# 7: 울산, 8: 세종, 9: 강원, 10: 충북, 11: 충남, 12: 전북,
# 13: 전남, 14: 경북, 15: 경남, 16: 제주

# 생애주기 순서 정의 및 라벨 부여
# 라벨 번호 기준:
# 0: 1.Single, 1: 2.가족형성기, 2: 3.자녀출산기,
# 3: 4.자녀성장기(1), 4: 5.자녀성장기(2),
# 5: 6.자녀출가기, 6: 7.노령
```

## 고유값이 1~4개인 피쳐 파악 ( 의미없어서 제거하려고 )

```python
def summarize_feature_groups(df, feature_groups):
    summary = {}

    total_assigned = 0
    total_existing = 0
    total_missing = 0

    print("📊 그룹별 피처 존재 여부 및 개수 요약:\n")
    for group_name, feature_list in feature_groups.items():
        existing = [col for col in feature_list if col in df.columns]
        missing = [col for col in feature_list if col not in df.columns]

        num_assigned = len(feature_list)
        num_existing = len(existing)
        num_missing = len(missing)

        summary[group_name] = {
            '총지정': num_assigned,
            '존재': num_existing,
            '누락': num_missing,
            '누락컬럼': missing
        }

        total_assigned += num_assigned
        total_existing += num_existing
        total_missing += num_missing

        print(f"📁 [{group_name}]")
        print(f"  - 지정된 피처 수: {num_assigned}")
        print(f"  - 존재하는 피처 수: {num_existing}")
        if missing:
            print(f"  - ⚠️ 누락된 피처 수: {num_missing} → {missing}")
        else:
            print(f"  - ✅ 모든 피처가 존재합니다.")
        print()

    print("📈 전체 요약:")
    print(f"  - 총 지정된 피처 수: {total_assigned}")
    print(f"  - 총 존재하는 피처 수: {total_existing}")
    print(f"  - 총 누락된 피처 수: {total_missing}")

    return summary

# 실행
group_summary = summarize_feature_groups(df_t1, feature_groups)
```
