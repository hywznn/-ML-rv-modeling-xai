## 스케일링 정규화 ( 이진피처는 제외 )

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler, MinMaxScaler, PowerTransformer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

# 컬럼 분류 결과 수동 정의
binary_features = [
    '회원여부_이용가능', '회원여부_이용가능_CA', '회원여부_이용가능_카드론', '회원여부_연체',
    '수신거부여부_TM', '수신거부여부_DM', '수신거부여부_메일', '수신거부여부_SMS',
    '동의여부_한도증액안내', '마케팅동의여부', '보유여부_해외겸용_본인', '이용가능여부_해외겸용_본인',
    '이용여부_3M_해외겸용_본인', '보유여부_해외겸용_신용_본인', '이용가능여부_해외겸용_신용_본인',
    '이용여부_3M_해외겸용_신용_본인', '카드론동의여부', 'RV전환가능여부', '일시불ONLY전환가능여부',
    '특별한도보유여부_R3M'
]

minmax_features = [
    '연령', 'CA이자율_할인전', 'CL이자율_할인전', 'RV일시불이자율_할인전',
    'RV현금서비스이자율_할인전', 'RV약정청구율', 'RV최소결제비율'
]

standard_features = [
    '_1순위카드이용건수', '_2순위카드이용건수', '카드신청건수', '한도심사요청건수',
    '한도요청거절건수', '자발한도감액횟수_R12M', '강제한도감액횟수_R12M', '한도증액횟수_R12M'
]

log_then_standard_features = [
    '이용금액_R3M_신용체크', '이용금액_R3M_신용', '이용금액_R3M_신용_가족', '이용금액_R3M_체크',
    '_1순위카드이용금액', '_2순위카드이용금액', '기본연회비_B0M', '제휴연회비_B0M',
    '할인금액_제휴연회비_B0M', '청구금액_기본연회비_B0M', '청구금액_제휴연회비_B0M',
    '최초한도금액', '카드이용한도금액', 'CA한도금액', '일시상환론한도금액', '월상환론한도금액',
    '자발한도감액금액_R12M', '강제한도감액금액_R12M', '한도증액금액_R12M',
    '상향가능한도금액', '상향가능CA한도금액', '월상환론상향가능한도금액'
]

# PowerTransformer 파이프라인 정의
power_transformer = Pipeline([
    ('yeo', PowerTransformer(method='yeo-johnson', standardize=True))
])

# ColumnTransformer 구성 (접두사 제거 옵션 추가)
preprocessor = ColumnTransformer(
    transformers=[
        ('minmax', MinMaxScaler(), minmax_features),
        ('standard', StandardScaler(), standard_features),
        ('power', power_transformer, log_then_standard_features),
    ],
    remainder='passthrough',
    verbose_feature_names_out=False  # 접두사 제거 옵션
)

# 구성 확인
preprocessor
```

```python
df_scaled = preprocessor.fit_transform(df_t2)

# 스케일링된 결과는 numpy 배열이므로, DataFrame으로 변환 (컬럼 이름 필요)
scaled_columns = preprocessor.get_feature_names_out()
df_scaled = pd.DataFrame(df_scaled, columns=scaled_columns)
```
